```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/amulya/Desktop/Capstone/DSCI-601-Amy/ProjectCode/Data Pipelines/Cleaned_AAPL.csv', parse_dates=['date'])
```

```python
# Calculate the 2-day SMA for the 'RET' column
df['SMA_RET'] = df['RET'].rolling(window=50).mean()
```

```python
print(df)
```

```python
plt.figure(figsize=(10, 5))  # Set the figure size
plt.plot(df['date'], df['RET'], label='RET', color='blue')  # Plot the original 'RET' values
plt.plot(df['date'], df['SMA_RET'], label='50-Day SMA of RET', color='red', linestyle='--')  # Plot the SMA values

plt.title('Simple Moving Average (SMA) vs. Original Returns')  # Set title
plt.xlabel('Date')  # Set x-axis label
plt.ylabel('Return')  # Set y-axis label
plt.legend()  # Show legend
plt.grid(True)  # Show grid
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping

plt.show()  # Display the plot
```
